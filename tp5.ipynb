```python
# Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Charger le jeu de données Iris
iris = load_iris()
X = iris.data

# Visualisation des deux premières caractéristiques (pour simplifier la visualisation)
plt.scatter(X[:, 0], X[:, 1], c='black', marker='o', s=50)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris dataset - Sepal length vs Sepal width')
plt.show()

# Utilisation de l'algorithme K-means pour regrouper les données en différents nombres de clusters
# pour trouver le nombre optimal de clusters avec la méthode du coude
sum_of_squared_distances = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sum_of_squared_distances.append(kmeans.inertia_)

# Tracer le graphique du nombre de clusters par rapport à la somme des carrés des distances
plt.plot(range(1, 11), sum_of_squared_distances, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Somme des carrés des distances')
plt.title('Méthode du coude pour choisir le nombre optimal de clusters')
plt.show()

# Utilisation de l'algorithme K-means pour regrouper les données en 3 clusters (le nombre de classes dans Iris)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Afficher les centres des clusters
print("Cluster centers:")
print(kmeans.cluster_centers_)

# Afficher les étiquettes de cluster attribuées à chaque exemple
print("Cluster labels:")
print(kmeans.labels_)

# Visualisation des clusters avec les centres
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, marker='o', s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris dataset with K-means clustering')
plt.legend()
plt.show()
```
